```python
from random import randrange


def cadre(mat):
    l,c = mat.shape
    for i in range(l):
        mat[i][1] = 0
        mat[i][c-2] = 0
    for j in range(c):
        mat[1][j] = 0
        mat[l-2][j] = 0
```

```python
c = np.ones((10,20))
cadre(c)
plt.imshow(c)
```

```python
tabrandom = np.random.randint(0, 21, (6,8), dtype = np.uint8)
plt.imshow(tabrandom)
;
```

```python
cadre(tabrandom)
plt.imshow(tabrandom)
;
```

cela ne va pas fonctionner pour un tableau de trop petite taille, qui ne possède pas plus de 2 lignes et colonnes. S'il en
possède moins de 4, la fonction n'a pas de sens.

```python
#On n'a pas le choix de la copie du tableau car on ne peut pas modifier ses dimensions.
def cadre2(tab):
    l,c = tab.shape
    res = np.zeros((l+2,c+2))
    for i in range(l):
        for j in range(c):
            res[i+1][j+1] = tab[i][j]
    plt.imshow(res)

cadre2(tabrandom)
#cela assure la correction de l'algorithme, on ne rencontrera pas de problème de typage ou de taille
```

```python
def anti_diagonal(lst):
    n = len(lst)
    res = np.zeros(n,n)
    for i in range(n):
        res[i,n-1-i] = lst[i]
    return(res)
```
